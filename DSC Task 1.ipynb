```python
df.columns=df.columns.str.replace(' ','_')
df.columns=df.columns.str.replace('1','Fir')
df.columns=df.columns.str.replace('%','Percent')
```

Replace characters so that we can perform operations using the column names

```python
mean=np.mean(df.First_Goal)
df.First_Goal=df.First_Goal.fillna(mean)
df.First_Goal=df.First_Goal.round(2)
```

Since most of the fields in the First Goal column have a not null value, it sounded right to fill the remaining null value (only 25%) of the total number of tuples with the average of the remaining 75%. Although we could as have filled the values by interpolating data from the previous values, some values were high (close to outlier values) when replaced.

```python
df=df.drop(df.columns[[0,25,26]], axis=1)
```

```python
df.Man_of_the_Match=df['Man_of_the_Match'].replace(['Yes', 'No'], ['1','0'])
```

```python
df.Goal_Scored=df.Goal_Scored.replace(df.Goal_Scored+df.Goals_in_PSO)
```

```python
f, ax = plt.subplots(figsize=(10, 8))
sb.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sb.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```

Visualing the said correlations with the help of a heatmap. The intensity allows us to visually interpret the correlation values.
As it can be seen the fields such Ball possesion percent, attempts,on target, off target ,blocked and corners are heavily dependent on each other. We can remove of these feature. 
We can select such feature on the basis of their effects on the other features.
For example,
Attempts and Corners and highly co related and so are attempts and blocked shots.
Blocked shots and corners too are co related but less when compared to attempts and blocked shots.
A change in the other two features can provide us with the change in the blocked shots but the vice versa being true gets covered by the fact that attempts and corners are more heavily related than blocked shots and corners.
Hence blocked shots can be removed
Similarly, we can find other fields to remove.
